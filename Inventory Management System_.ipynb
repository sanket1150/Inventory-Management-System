```python
def customer():
    print("1. show menu")
    print("2. About Us")

    key = record.keys()
    use  = int(input("enter your choice : "))
    if use == 1 :
        
        print("products available in inventory is :")
        for i in key:
            print(i,record[str(i)]["name"])
            
        
    elif use == 2 :
        print("**ETG**\n","The Inventory Management System")
            
    else:
        print("please! enter correct choice")
    
    product  = int(input("No. of items you want : "))
   
    trans = 1
    a = 0
    b = 0
    for i in range(product):
        pro_id = str(input("enter product id you want to purchase : "))
    
    
        quantity = int(input("enter net quantity : "))
        if quantity <0 and quantity > record[pro_id]["Available Quantity"] :
            print ("not available")
    
        print("your selected product is : ")
    
        pro_name = (pro_id,record[str(pro_id)]["name"])
        prod_id  = (record[str(pro_id)]["name"])
        print(pro_name)
    
        a = a + record[pro_id]["price"] * quantity
    
        b = b + record[pro_id]["MRP."] * quantity
        used_quantity = record[pro_id]["Available Quantity"] - quantity
        record[pro_id]["Available Quantity"] = used_quantity 
        
    detail = {(trans + 1):{"Time" : atime,"Name" : prod_id , "Quantity" : quantity,  "Amount Paid" :a ,}}

    sd = json.dumps(detail)
    fs = open('sales.json', 'a')
    fs.write(sd)
    fs.close()
    
    
    print("   **********Total MRP**********\n  ","          ",b) 

    print("**********Discounted Amount**********\n : ","          ",a)

    print ("   **********you saved**********\n ","      ",b-a,"on MRP.")
    print ("*****Thank you ! Visit Again have a Good Day*****")
```
